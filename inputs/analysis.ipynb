```python
import numpy as np
import json
```

```python
json_file = '../outputs/thermoout.json'
stream =  open(json_file)
data = json.load(stream)
```

```python
msfl = []
gas_ideal = []
FM3M = []
P3c1 = []
Li2BeF4 = []
NdF3 = []
temperature = []
pressure = []

```

```python
data["1"].keys()
```

```python
for key in list(data.keys()):
  msfl.append(data[key]['solution phases']["MSFL"]["moles"])
  gas_ideal.append(data[key]['solution phases']["gas_ideal"]["moles"])
  FM3M.append(data[key]['solution phases']["FM3M"]["moles"])
  P3c1.append(data[key]['solution phases']["P3c1"]["moles"])
  Li2BeF4.append(data[key]['pure condensed phases']["Li2BeF4_S1(s)"]["moles"])
  NdF3.append(data[key]['pure condensed phases']["NdF3_S1(s)"]["moles"])
  temperature.append(data[key]['temperature'])
  pressure.append(data[key]['pressure'])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(temperature, msfl, 'r-o', label='MSFL')
plt.plot(temperature, gas_ideal, 'b-*', label='Ideal Gas')
plt.plot(temperature, P3c1, 'g-s', label='P3C1')
# plt.plot(temperature, Li2BeF4, 'c--o', label='Li2BeF4_S1(s)')
# plt.plot(temperature, NdF3, 'm--*', label='NdF3_S1(s)')
plt.legend(loc='upper left')
plt.xlabel("Temperature [K]")
plt.ylabel("Phase [moles]")
plt.yscale("log")
plt.show()
```
