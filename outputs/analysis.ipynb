```python
T = []
G = []
x_Cr = []
x_Ni = []
x_Fe = []
x_Li = []
x_Be = []
x_F = []
moles = []
x_LiF = []
x_BeF2 = []
x_BeF4 = []

for key in data.keys():
  T.append(data[key]['temperature'])
  G.append(data[key]['integral Gibbs energy'])
  try:
    x_Cr.append(data[key]['elements']['Cr']['moles'])
  except:
    x_Cr.append(0.0)
  try:
    x_Ni.append(data[key]['elements']['Ni']['moles'])
  except:
    x_Ni.append(0.0)
  try:
    x_Fe.append(data[key]['elements']['Fe']['moles'])
  except:
    x_Fe.append(0.0)
  try:
    x_Li.append(data[key]['elements']['Li']['moles'])
  except:
    x_Li.append(0.0)
  try:
    x_Be.append(data[key]['elements']['Be']['moles'])
  except:
    x_Be.append(0.0)
  try:
    x_F.append(data[key]['elements']['F']['moles'])
  except:
    x_F.append(0.0)
  moles.append(data[key]['solution phases']['MSFL']['moles of endmembers'])
  x_LiF.append(data[key]['solution phases']['MSFL']['endmembers']['LiF']['mole fraction'])
  x_BeF2.append(data[key]['solution phases']['MSFL']['endmembers']['Be[CN=IV]F2']['mole fraction'])
  x_BeF4.append(data[key]['solution phases']['MSFL']['endmembers']['Be[Dimer]F4']['mole fraction'])

T = np.array(T)
G = np.array(G)
n_elements = np.array(x_Cr) + np.array(x_Fe) + np.array(x_Ni) + np.array(x_F) + np.array(x_Li) +  np.array(x_Be)
G_m = G / n_elements
x_Cr = np.array(x_Cr) /  n_elements
x_Fe = np.array(x_Fe) /  n_elements
x_Ni = np.array(x_Ni) /  n_elements
x_Li = np.array(x_Li) /  n_elements
x_Be = np.array(x_Be) /  n_elements
x_F = np.array(x_F) /  n_elements
moles_endmembers = np.array(moles)
x_LiF = np.array(x_LiF)
x_BeF2 = np.array(x_BeF2)
x_BeF4 = np.array(x_BeF4)
```

```python
g0_LiF  = -617789.909 + 386.915780 * T - 64.1830000 * T * np.log(T)
g0_BeF2 = -972101.922 + 400.891551 * T - 66.5685000 * T * np.log(T) - 0.790250000E-02 * T**2 - 36975.7500 / T
g0_BeF4 = -2094203.84 + 801.783103 * T - 133.137000 * T * np.log(T) - 0.158050000e-01 * T**2 - 73951.5000 / T
g_CrF2 = (G - moles_endmembers * (g0_LiF * x_LiF + g0_BeF2 * x_BeF2 + g0_BeF4 * x_BeF4)) / 9.64853321233100184e4 - 2.871 * 2 * (x_Cr + x_Fe + x_Ni)
```
